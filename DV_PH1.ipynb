## Dataset Description

The dataset used in this project is the "Supermarket Sales Dataset" obtained from Kaggle.

This dataset contains approximately 1000 records and 17 features describing sales transactions in a supermarket across different cities.

### Dataset Structure:
- Rows: ~1000 transactions
- Columns: 17 variables

### Main Variables:
- City: The city where the transaction took place
- Customer type: Member or Normal customer
- Gender: Customer gender
- Product line: Category of the product
- Unit price: Price of a single product
- Quantity: Number of products purchased
- Sales: Total sales amount for the transaction
- Date: Date of purchase
- Time: Time of purchase
- Payment: Payment method used

### Data Types:
- Numerical variables: Unit price, Quantity, Sales, Tax, Income
- Categorical variables: City, Gender, Product line, Payment
- Date/Time variables: Date and Time

### Purpose of the Dataset:
This dataset is suitable for data visualization and analysis because it contains a mix of numerical and categorical variables. It allows the exploration of sales trends, customer behavior, and product performance using different visualization techniques.

The dataset was selected because it provides meaningful insights and supports various types of charts required in this project.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\Dell\Desktop\DV\SuperMarket Analysis.csv")
df.head()
```

```python
df.shape
df.columns
df.info()
```

```python
df.head()
df.isnull().sum()
df.duplicated().sum()
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()
```

```python
df['Sales'] = df['Sales'].astype(float)
```

```python
df.info()
```

```python
df.fillna(0, inplace=True)
df = df.drop_duplicates()

df.isnull().sum()
df.duplicated().sum()
```

## Matplotlib Visualizations

```python
df.groupby('Date')['Sales'].sum().plot()
plt.title("Sales Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales")
plt.grid()
plt.show()
```

This chart shows how sales change over time. It helps identify trends and patterns in sales performance.

```python
df.groupby('City')['Sales'].sum().plot(kind='bar')
plt.title("Sales by City")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.show()
```

This chart shows the total sales for each city. It helps identify which city generates the highest revenue.

```python
plt.scatter(df['Unit price'], df['Quantity'])
plt.title("Unit Price vs Quantity")
plt.xlabel("Unit Price")
plt.ylabel("Quantity")
plt.show()
```

This scatter plot shows the relationship between unit price and quantity purchased.

```python
df['Sales'].hist()
plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")
plt.show()
```

This histogram shows the distribution of sales values and how frequently they occur.

```python
df['Payment'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Payment Methods Distribution")
plt.ylabel("")
plt.show()
```

This chart shows the distribution of different payment methods used by customers.

```python
fig, ax = plt.subplots(1, 2, figsize=(12,5))

df['Sales'].hist(ax=ax[0])
ax[0].set_title("Sales Distribution")

df.groupby('City')['Sales'].sum().plot(kind='bar', ax=ax[1])
ax[1].set_title("Sales by City")

plt.tight_layout()
plt.show()
```

This subplot combines two visualizations to compare sales distribution and sales across cities.

## Seaborn Visualizations

```python
sns.boxplot(x='Product line', y='Sales', data=df)
plt.xticks(rotation=45)
plt.title("Sales by Product Line")
plt.show()
```

This boxplot shows the distribution of sales across different product lines and helps identify variations and outliers.

sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()

The heatmap shows the correlation between numerical variables, helping identify strong relationships between features.

```python
sns.countplot(x='City', data=df)
plt.title("Transactions per City")
plt.show()
```

This chart shows the number of transactions in each city, indicating customer activity levels.

```python
sns.pairplot(df)
plt.show()
```

The pairplot visualizes relationships between multiple numerical variables and helps identify patterns and correlations.
